```python
import re

def cleans(text):
    text = text.lower()
    text = re.sub("http\S+\s", " ", text)
    text = re.sub("RT|CC", " ", text)
    text = re.sub("#\S+\s", " ", text)
    text = re.sub("@\S+", " ", text)
    text = re.sub("[%s]" % re.escape("""!"#&%&"()*+,-./:;<=>?@[\]^_'{|}~"""), " ", text)
    text = re.sub(r"[^\x00-\x7f]", " ", text)  # Fixed line
    text = re.sub("\s+", " ", text)  # Also better to remove extra spaces
    return text

df_train["text"] = df_train["text"].apply(lambda x: cleans(x))
```

```python
import re

def cleans(text):
    text = text.lower()
    text = re.sub("http\S+\s", " ", text)
    text = re.sub("RT|CC", " ", text)
    text = re.sub("#\S+\s", " ", text)
    text = re.sub("@\S+", " ", text)
    text = re.sub("[%s]" % re.escape("""!"#&%&"()*+,-./:;<=>?@[\]^_'{|}~"""), " ", text)
    text = re.sub(r"[^\x00-\x7f]", " ", text)  # Fixed line
    text = re.sub("\s+", " ", text)  # Also better to remove extra spaces
    return text

df_test["text"] = df_test["text"].apply(lambda x: cleans(x))
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lab = LabelEncoder()
```

```python
df_train["sentiment"] = lab.fit_transform(df_train["sentiment"])
```

```python
df_train.sample(10)
```

```python
df_test["sentiment"] = lab.fit_transform(df_test["sentiment"])
```

```python
df_test.head()
```

```python
df = pd.concat([df_train,df_test],axis = 0)
```

```python
df.shape
```

```python
df.sample(frac = True,ignore_index = True)
```

```python
df.head()
```

```python
x = df["text"]
y = df["sentiment"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words = "english")
```

```python
x_train = tfidf.fit_transform(x_train)
x_test = tfidf.transform(x_test)
```

```python
print(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(max_iter=1000)
```

```python
lg.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = lg.predict(x_test)
print("Accuracy_score =",accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
print("Accuracy_score =",accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rd = RandomForestClassifier()
```

```python
rd.fit(x_train,y_train)
```

```python
y_pred = rd.predict(x_test)
print("Accuracy_score =",accuracy_score(y_test,y_pred))
```

```python
from sklearn.svm import SVC
svc = SVC()
```

```python
svc.fit(x_train,y_train)
```

```python
y_pred = svc.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
print("Accuracy_score = ",accuracy_score(y_test,y_pred))
```

```python
ada.fit(x_train,y_train)
```

```python
y_pred = ada.predict(x_test)
print("Accuracy_score =",accuracy_score(y_test,y_pred))
```

```python
params = [("lg",lg),("dt",dt),("rd",rd),("svc",svc),("knn",knn),("xgb",xgb),("ada",ada),("gd",gd),("bag",bag)]
```

```python
from sklearn.ensemble import VotingClassifier
vt = VotingClassifier(estimators = params)
```
